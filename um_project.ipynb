```python
import pandas as pd
import glob
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
```

```python
neutrinoFiles = glob.glob("atm_neutrino_class*")
neutrinoFiles.sort()
neutrinoFiles
```

```python
dataFrames = []
for filePath in neutrinoFiles:
  dataFrames.append(pd.read_hdf(filePath, "y"))

df = pd.concat(dataFrames)

df['isNeutrino'] = 1
df.describe()
```

```python
df
```

```python
df2 = pd.read_hdf("atm_muon.h5", "y")
df2['isNeutrino'] = 0
df2.describe()
```

```python
df = pd.concat([df, df2])
df.describe()
```

```python
df.count()
```

```python
df = df.dropna()
```

```python
train, test = train_test_split(df, test_size=0.33, random_state=2137, stratify=df['isNeutrino'])

print(f'Train set length: {len(train)}, Test set length: {len(test)}')
```

```python
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
clf.fit(train.drop(columns=['isNeutrino']), train['isNeutrino'])
```

```python
clf.score(train.drop(columns=['isNeutrino']), train['isNeutrino'])
```

```python
clf.score(test.drop(columns=['isNeutrino']), test['isNeutrino'])
```
